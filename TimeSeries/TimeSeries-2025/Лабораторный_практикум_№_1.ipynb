```python
from datetime import datetime, timedelta

# –°–ª–æ–≤–∞—Ä—å –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –Ω–æ–º–µ—Ä–æ–≤ –º–µ—Å—è—Ü–µ–≤ –≤ —Ä—É—Å—Å–∫–∏–µ –Ω–∞–∑–≤–∞–Ω–∏—è
month_names = {
    1: "—è–Ω–≤–∞—Ä—è", 2: "—Ñ–µ–≤—Ä–∞–ª—è", 3: "–º–∞—Ä—Ç–∞", 4: "–∞–ø—Ä–µ–ª—è",
    5: "–º–∞—è", 6: "–∏—é–Ω—è", 7: "–∏—é–ª—è", 8: "–∞–≤–≥—É—Å—Ç–∞",
    9: "—Å–µ–Ω—Ç—è–±—Ä—è", 10: "–æ–∫—Ç—è–±—Ä—è", 11: "–Ω–æ—è–±—Ä—è", 12: "–¥–µ–∫–∞–±—Ä—è"
}

start_date = datetime(2023, 10, 1)
end_date = datetime(2025, 10, 1)

date_list = []
current_date = start_date

while current_date <= end_date:
    day = current_date.day
    month_ru = month_names[current_date.month]
    year = current_date.year

    date_str = f"{day}-{month_ru}#{year}"
    date_list.append(date_str)

    current_date += timedelta(days=1)

print(date_list)
```

- –£–¥–∞–ª–∏—Ç—å –¥—É–±–ª–∏–∫–∞—Ç—ã –ø–æ –≤—Ä–µ–º–µ–Ω–∏

  - –ú–µ—Ç–æ–¥ IQR (–º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π —Ä–∞–∑–º–∞—Ö)
  - Z-score (–µ—Å–ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±–ª–∏–∑–∫–æ –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É)

> üõ†Ô∏è **–ò–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç—ã:** `pandas`, `numpy`, `pytz`


```python
import pandas as pd
import pytz
from datetime import timedelta

# –ó–∞–º–µ–Ω–∞ '24:00' —Å '00:00' –∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞—è –Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –¥–∞—Ç—ã (–£–≤–µ–ª–∏—á–µ–Ω–∏–µ –¥–æ —Ç–µ–∫—É—â–µ–≥–æ –≥–æ–¥–∞)
df_weather['–î–∞—Ç–∞'] = df_weather.apply(lambda row: (pd.to_datetime(row['–î–∞—Ç–∞']) + timedelta(days=1)).strftime('%Y-%m-%d') if row['–í—Ä–µ–º—è'] == '24:00' else row['–î–∞—Ç–∞'], axis=1)
df_weather['–í—Ä–µ–º—è'] = df_weather['–í—Ä–µ–º—è'].replace('24:00', '00:00')

# –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞–Ω–Ω—ã—Ö
df_weather['–í–µ—Ç–µ—Ä —Å–∫–æ—Ä–æ—Å—Ç—å, –º/—Å']=df_weather['–í–µ—Ç–µ—Ä —Å–∫–æ—Ä–æ—Å—Ç—å, –º/—Å'].fillna(0.0)

# –ö–æ–º–±–∏–Ω–∞—Ü–∏—è '–î–∞—Ç–∞' –∏'–í—Ä–µ–º—è' –∫–æ–ª–æ–Ω–æ–∫ –∏ –ø—Ä–µ–≤—Ä–∞—â–µ–Ω–∏–µ –≤ datetime –æ–±—ä–µ–∫—Ç
df_weather['timestamp'] = pd.to_datetime(df_weather['–î–∞—Ç–∞'] + ' ' + df_weather['–í—Ä–µ–º—è'], format='%Y-%m-%d %H:%M')

# –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞—Ü–∏–∏ –ø–æ timestamp
df_weather = df_weather.set_index('timestamp')

# –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–µ timezone Europe/Moscow –¥–ª—è –∏–Ω–¥–µ–∫—Å–∞
moscow_tz = pytz.timezone('Europe/Moscow')

# –ü—Ä–æ–≤–µ—Ä–∫–∞, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –ª–∏ –∏–Ω–¥–µ–∫—Å —É–∂–µ —á–∞—Å–æ–≤–æ–º—É –ø–æ—è—Å—É
if df_weather.index.tz is None:
    df_weather.index = df_weather.index.tz_localize(pytz.utc).tz_convert(moscow_tz)
else:
    df_weather.index = df_weather.index.tz_convert(moscow_tz)

# –£–¥–∞–ª–µ–Ω–∏–µ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –Ω–∞ –æ—Å–Ω–æ–≤–µ –∏–Ω–¥–µ–∫—Å–∞ (–≤—Ä–µ–º–µ–Ω–Ω–æ–π –º–µ—Ç–∫–∏)
df_weather = df_weather[~df_weather.index.duplicated(keep='first')]

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ—Å—Ç–∏
if not df_weather.index.is_monotonic_increasing:
    print("Warning: The time series is not monotonic. Sorting by index.")
    df_weather = df_weather.sort_index()
else:
    print("The time series is monotonic.")

# –í—ã–≤–æ–¥ –≤ –∫–æ–Ω—Å–æ–ª—å –¥–∞—Ç–∞—Å–µ—Ç–∞ –∏ –∏–Ω—Ñ—ã –ø–æ –Ω–µ–º—É
display(df_weather.head())
display(df_weather.info())
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ '–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞' –≤ —á–∏—Å–ª–æ–≤–æ–π —Ñ–æ—Ä–º–∞—Ç
df_weather['–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞'] = df_weather['–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞'].str.replace('¬∞', '').str.replace('+', '')
df_weather['–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞'] = df_weather['–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞'].astype(float)

# –í—ã–≤–æ–¥ –ø–µ—Ä–≤—ã—Ö –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Å—Ç—Ä–æ–∫ –∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ DataFrame –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏
display(df_weather.head())
display(df_weather.info())
```

```python
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –æ–ø–∏—Å–∞—Ç–µ–ª—å–Ω–æ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
print("Descriptive Statistics:")
display(df_weather.describe())

# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
import matplotlib.pyplot as plt

numerical_cols = df_weather.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols), 1, i + 1)
    plt.plot(df_weather.index, df_weather[col])
    plt.title(col)
plt.tight_layout()
plt.show()

# –ì—Ä–∞—Ñ–∏–∫ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols), 1, i + 1)
    df_weather[col].hist(bins=30)
    plt.title(col)
# –ê–≤—Ç–æ–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∫–∞ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∞–º–∏
plt.tight_layout()
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω—ã—Ö –¥–∏–∞–≥—Ä–∞–º–º –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols), 1, i + 1)
    df_weather.boxplot(column=col)
    plt.title(col)
plt.tight_layout()
plt.show()

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã
import seaborn as sns

correlation_matrix = df_weather[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

–î–∞–ª–µ–µ –º—ã –æ–ø—Ä–µ–¥–µ–ª–∏–º –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –≤ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö, –∏—Å–ø–æ–ª—å–∑—É—è –º–µ—Ç–æ–¥ –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω–æ–≥–æ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ (IQR).

```python
import numpy as np

# –í—ã–±–æ—Ä —Ç–æ–ª—å–∫–æ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –æ–±–Ω–∞—Ä—É–∂–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤
numerical_cols = df_weather.select_dtypes(include=['float64', 'int64']).columns

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ IQR –¥–ª—è –∫–∞–∂–¥ —á–∏—Å–ª–æ–≤–æ–π –∫–æ–ª–æ–Ω–∫–∏
Q1 = df_weather[numerical_cols].quantile(0.25)
Q3 = df_weather[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≥—Ä–∞–Ω–∏—Ü –¥–ª—è –≤—ã–±—Ä–æ—Å–æ–≤
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# –í—ã—è–≤–ª–µ–Ω–∏–µ –≤—ã–±—Ä–æ—Å–æ–≤
outliers = df_weather[numerical_cols][
    ((df_weather[numerical_cols] < lower_bound) | (df_weather[numerical_cols] > upper_bound))
]

print("–í—ã—è–≤–ª–µ–Ω–Ω—ã–µ –≤—ã–±—Ä–æ—Å—ã (NaN —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –≤—ã–±—Ä–æ—Å–∞ –≤ —ç—Ç–æ–π –ø–æ–∑–∏—Ü–∏–∏):")
display(outliers.head())

# –í—ã –º–æ–∂–µ—Ç–µ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∏–ª–∏ –æ–±—Ä–∞–±–æ—Ç–∞—Ç—å —ç—Ç–∏ –≤—ã–±—Ä–æ—Å—ã (–Ω–∞–ø—Ä–∏–º–µ—Ä, –æ–≥—Ä–∞–Ω–∏—á–∏—Ç—å –∏—Ö, —É–¥–∞–ª–∏—Ç—å —Å—Ç—Ä–æ–∫–∏ –∏–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥—Ä—É–≥–æ–π –º–µ—Ç–æ–¥)
# –ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ:
print("\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±—Ä–æ—Å–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü–µ:")
display(outliers.count())
```

–¢–µ–ø–µ—Ä—å, –∫–æ–≥–¥–∞ –Ω–∞—á–∞–ª—å–Ω–∞—è –æ—á–∏—Å—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∑–∞–≤–µ—Ä—à–µ–Ω–∞, –æ–±—Ä–∞—Ç–∏–º—Å—è –∫ –æ—Ç—Å—É—Ç—Å—Ç–≤—É—é—â–∏–º –∑–Ω–∞—á–µ–Ω–∏—è–º. –ú—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—é, —á—Ç–æ–±—ã –∑–∞–ø–æ–ª–Ω–∏—Ç—å –Ω–µ–¥–æ—Å—Ç–∞—é—â–∏–µ —Ç–æ—á–∫–∏ –¥–∞–Ω–Ω—ã—Ö –≤ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö.

```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞, –Ω–µ—Ç –ª–∏ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–µ—Ä–µ–¥ –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–µ–π
print("–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–µ—Ä–µ–¥ –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–µ–π:")
display(df_weather.isnull().sum())

# –í—ã–±–æ—Ä —Ç–æ–ª—å–∫–æ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏, –∏—Å–ø–æ–ª—å–∑—É—è .loc, –≤–æ –∏–∑–±–µ–∂–∞–Ω–∏–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ —Å –ø–æ–º–æ—â—å—é copywith Warning
numerical_cols = df_weather.select_dtypes(include=['float64', 'int64']).columns
df_weather.loc[:, numerical_cols] = df_weather.loc[:, numerical_cols].interpolate(method='time')

# –ü—Ä–æ–≤–µ—Ä–∫–∞, –Ω–µ—Ç –ª–∏ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—Å–ª–µ –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏
print("\n–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ—Å–ª–µ –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏:")
display(df_weather.isnull().sum())

# Display info to see non-null counts
display(df_weather.info())
```

### **3.4. –≠—Ç–∞–ø 4. –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ—Å—Ç—å –∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Ç–µ—Å—Ç—ã**

**–ó–∞–¥–∞—á–∞:** –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Ä—è–¥ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω—ã–º ‚Äî –∫–ª—é—á–µ–≤–æ–µ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏–µ –¥–ª—è –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏—Ö –º–µ—Ç–æ–¥–æ–≤ –∞–Ω–∞–ª–∏–∑–∞.

#### üìå –ß—Ç–æ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å:
- –í–∏–∑—É–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑: –µ—Å—Ç—å –ª–∏ —Ç—Ä–µ–Ω–¥, –º–µ–Ω—è–µ—Ç—Å—è –ª–∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤–æ –≤—Ä–µ–º–µ–Ω–∏?
- –†–∞—Å—Å—á–∏—Ç–∞—Ç—å —Å–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ä–µ–¥–Ω–µ–µ –∏ —Å–∫–æ–ª—å–∑—è—â—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é (–ø–æ –æ–∫–Ω–∞–º 30, 60, 90 —Ç–æ—á–µ–∫) ‚Äî —Å—Ç–∞–±–∏–ª—å–Ω—ã –ª–∏ –æ–Ω–∏?
- –ü—Ä–∏–º–µ–Ω–∏—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Ç–µ—Å—Ç—ã:
  - **–¢–µ—Å—Ç –î–∏–∫–∏-–§—É–ª–ª–µ—Ä–∞ (ADF)** ‚Äî –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Ä—è–¥ –∏–º–µ–µ—Ç –µ–¥–∏–Ω–∏—á–Ω—ã–π –∫–æ—Ä–µ–Ω—å (–Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω)
  - **–¢–µ—Å—Ç KPSS** ‚Äî –Ω—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Ä—è–¥ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω
- –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å p-value:
  - ADF: p < 0.05 ‚Üí —Ä—è–¥ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω
  - KPSS: p > 0.05 ‚Üí —Ä—è–¥ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω
- –ï—Å–ª–∏ —Ä—è–¥ –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω ‚Äî –ø—Ä–∏–º–µ–Ω–∏—Ç—å **–¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏—Ä–æ–≤–∞–Ω–∏–µ 1-–≥–æ –ø–æ—Ä—è–¥–∫–∞** –∏ –ø–æ–≤—Ç–æ—Ä–∏—Ç—å —Ç–µ—Å—Ç—ã

> üìö **–ë–∏–±–ª–∏–æ—Ç–µ–∫–∏:** `statsmodels.tsa.stattools.adfuller`, `kpss`

```python
from statsmodels.tsa.stattools import adfuller, kpss
import pandas as pd

# –í—ã–±–æ—Ä —Ç–æ–ª—å–∫–æ —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ—Å—Ç–∏
numerical_cols = df_weather.select_dtypes(include=['float64', 'int64']).columns

# –ü—Ä–æ–≤–µ—Ä–∫–∞ ADF –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–æ–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
print("–î–æ–ø–æ–ª–Ω–µ–Ω–Ω—ã–π —Ç–µ—Å—Ç –î–∏–∫–∏-–§—É–ª–ª–µ—Ä–∞ (ADF):")
for col in numerical_cols:
    result = adfuller(df_weather[col].dropna()) # Drop NaNs just in case, although interpolation was done
    print(f"Column: {col}")
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"  {key}: {value}")
    if result[1] <= 0.05:
        print("–í—ã–≤–æ–¥: –°–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —Ä—è–¥ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")
    else:
        print("–í—ã–≤–æ–¥: C–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —Ä—è–¥ –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")
        kk = 0
        while result[1]>0.05 or kk<3:
            df_weather[col] = df_weather[col].diff().fillna(0)
            result = adfuller(df_weather[col].dropna())
            kk+=1
        if result[1]<=0.05:
            print(f'Column: {col} —Ç–µ–ø–µ—Ä—å —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–∞!')
    print("-" * 30)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ KPSS –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —á–∏—Å–ª–æ–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
print("\n–¢–µ—Å—Ç –ö–≤—è—Ç–∫–æ–≤—Å–∫–æ–≥–æ-–§–∏–ª–ª–∏–ø—Å–∞-–®–º–∏–¥—Ç–∞-–®–∏–Ω–∞ (KPSS):")
for col in numerical_cols:
    # KPSS —Ç–µ—Å—Ç –º–æ–∂–µ—Ç –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Å—Ç–æ–ª–±–µ—Ü –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ
    result = kpss(df_weather[col].dropna(), regression='c') # "c" –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ —Ç—Ä–µ–Ω–¥–∞, "ct" –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç –Ω–∞–ª–∏—á–∏–µ —Ç—Ä–µ–Ω–¥–∞
    print(f"Column: {col}")
    print(f"KPSS Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print("Critical Values:")
    for key, value in result[3].items():
        print(f"  {key}: {value}")
    if result[1] >= 0.05:
        print("–í—ã–≤–æ–¥: –†—è–¥, —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω—ã–º.")
    else:
        print("–í—ã–≤–æ–¥: –†—è–¥, —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")
    print("-" * 30)
```

### **3.5. –≠—Ç–∞–ø 5. –°–æ–∑–¥–∞–Ω–∏–µ –ª–∞–≥–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ —Å–∫–æ–ª—å–∑—è—â–∏—Ö —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫**

**–ó–∞–¥–∞—á–∞:** –ü–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–µ–π.

#### ‚è≥ –ß—Ç–æ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å:
- –°–æ–∑–¥–∞—Ç—å –ª–∞–≥–∏ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π: `target_lag_1`, `target_lag_7`, `target_lag_30`
- –°–æ–∑–¥–∞—Ç—å –ª–∞–≥–∏ –¥–ª—è –∑–Ω–∞—á–∏–º—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–µ—Å–ª–∏ –µ—Å—Ç—å –≥–∏–ø–æ—Ç–µ–∑–∞ –æ –∑–∞–ø–∞–∑–¥—ã–≤–∞–Ω–∏–∏ –≤–ª–∏—è–Ω–∏—è)
- –†–∞—Å—Å—á–∏—Ç–∞—Ç—å —Å–∫–æ–ª—å–∑—è—â–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏:
  - –°–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ä–µ–¥–Ω–µ–µ: `target_rolling_mean_7`, `target_rolling_mean_30`
  - –°–∫–æ–ª—å–∑—è—â–µ–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: `target_rolling_std_7`
- –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –º–µ–∂–¥—É –ª–∞–≥–∞–º–∏ –∏ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ‚Äî –∫–∞–∫–∏–µ –ª–∞–≥–∏ –Ω–∞–∏–±–æ–ª–µ–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã?
- –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç—å –ø–æ—Å–ª–µ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –ª–∞–≥–æ–≤

> üí° **–°–æ–≤–µ—Ç:** –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ `df.shift()` –∏ `df.rolling()` –≤ pandas

```python
# –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞—è '–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª' —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è, –æ—Å–Ω–æ–≤–∞–Ω–Ω–∞—è –Ω–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–º –∞–Ω–∞–ª–∏–∑–µ
target_col = '–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª'

# –°–æ–∑–¥–∞–Ω–∏–µ –∑–∞–ø–∞–∑–¥—ã–≤–∞—é—â–∏—Ö –æ–±—ä–µ–∫—Ç–æ–≤ –¥–ª—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
df_weather[f'{target_col}_lag_1'] = df_weather[target_col].shift(1)
df_weather[f'{target_col}_lag_7'] = df_weather[target_col].shift(7)
df_weather[f'{target_col}_lag_30'] = df_weather[target_col].shift(30)

# –°–æ–∑–¥–∞–Ω–∏–µ –æ–±—ä–µ–∫—Ç–æ–≤ —Å –∑–∞–¥–µ—Ä–∂–∫–æ–π –¥–ª—è –¥—Ä—É–≥–∏—Ö –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ –∑–Ω–∞—á–∏–º—ã—Ö —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤
# –°—É–¥—è –ø–æ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏, "–ß–∏—Å–ª–æ —Ä–æ–¥–∏–≤—à–∏—Ö—Å—è, —á–µ–ª" –∏ "–ß–∏—Å–ª–æ —É–º–µ—Ä—à–∏–π, —á–µ–ª" —Å–∏–ª—å–Ω–æ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—Ç —Å —Ü–µ–ª—å—é.
# –°–æ–∑–¥–∞–¥–∏–º –ª–∞–≥–∏ –∏ –¥–ª—è –Ω–∏—Ö —Ç–æ–∂–µ.
other_cols_to_lag = ['–ß–∏—Å–ª–æ —Ä–æ–¥–∏–≤—à–∏—Ö—Å—è, —á–µ–ª', '–ß–∏—Å–ª–æ —É–º–µ—Ä—à–∏—Ö, —á–µ–ª']
for col in other_cols_to_lag:
    df_weather[f'{col}_lag_1'] = df_weather[col].shift(1)
    df_weather[f'{col}_lag_7'] = df_weather[col].shift(7)
    df_weather[f'{col}_lag_30'] = df_weather[col].shift(30)

# –í—ã—á–∏—Å–ª–∏—Ç–µ —Å–∫–æ–ª—å–∑—è—â—É—é —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
df_weather[f'{target_col}_rolling_mean_7'] = df_weather[target_col].rolling(window=7).mean()
df_weather[f'{target_col}_rolling_mean_30'] = df_weather[target_col].rolling(window=30).mean()
df_weather[f'{target_col}_rolling_std_7'] = df_weather[target_col].rolling(window=7).std()

# –ü–µ—Ä–≤—ã–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫ –¥–ª—è –æ–∑–Ω–∞–∫–æ–º–ª–µ–Ω–∏—è —Å —Ñ—É–Ω–∫—Ü–∏—è–º–∏
display(df_weather.head())

# –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏–Ω—Ñ—ã –æ —Ñ—Ä–µ–π–º–µ –¥–∞–Ω–Ω—ã—Ö, —á—Ç–æ–± —É–≤–∏–¥–µ—Ç—å –Ω–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã
display(df_weather.info())
```

### **3.6. –≠—Ç–∞–ø 6. –ê–Ω–∞–ª–∏–∑ –∞–≤—Ç–æ–∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏: ACF –∏ PACF**

**–ó–∞–¥–∞—á–∞:** –í—ã—è–≤–∏—Ç—å —Å—Ç—Ä—É–∫—Ç—É—Ä—É –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Ä—è–¥–∞ –æ—Ç —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–æ—à–ª–æ–≥–æ.

#### üìà –ß—Ç–æ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å:
- –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å:
  - **ACF (Autocorrelation Function)** ‚Äî –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Ä—è–¥–∞ —Å –µ–≥–æ –ª–∞–≥–∞–º–∏ (–≤–∫–ª—é—á–∞—è –∫–æ—Å–≤–µ–Ω–Ω—ã–µ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏)
  - **PACF (Partial Autocorrelation Function)** ‚Äî ‚Äú—á–∏—Å—Ç–∞—è‚Äù –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å –ª–∞–≥–æ–º, –∏—Å–∫–ª—é—á–∞—è –≤–ª–∏—è–Ω–∏–µ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã—Ö –ª–∞–≥–æ–≤
- –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏:
  - –†–µ–∑–∫–∏–π –æ–±—Ä—ã–≤ –≤ PACF –Ω–∞ –ª–∞–≥–µ *p* ‚Üí –≤–æ–∑–º–æ–∂–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ AR(p)
  - –ü–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ–µ –∑–∞—Ç—É—Ö–∞–Ω–∏–µ –≤ ACF ‚Üí –≤–æ–∑–º–æ–∂–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ MA(q)
- –£–∫–∞–∑–∞—Ç—å, –∫–∞–∫–∏–µ –ª–∞–≥–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã (–≤—ã—Ö–æ–¥—è—Ç –∑–∞ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª)

> üìä **–ë–∏–±–ª–∏–æ—Ç–µ–∫–∏:** `statsmodels.graphics.tsaplots.plot_acf`, `plot_pacf`

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞—è, —á—Ç–æ '–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª' —è–≤–ª—è–µ—Ç—Å—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
target_col = '–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª'

# –ì—Ä–∞—Ñ–∏–∫ ACF
plt.figure(figsize=(12, 6))
plot_acf(df_weather[target_col].dropna(), lags=40, ax=plt.gca())
plt.title(f'Autocorrelation Function (ACF) for {target_col}')
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.show()

# –ì—Ä–∞—Ñ–∏–∫ PACF
plt.figure(figsize=(12, 6))
plot_pacf(df_weather[target_col].dropna(), lags=40, ax=plt.gca())
plt.title(f'Partial Autocorrelation Function (PACF) for {target_col}')
plt.xlabel('Lags')
plt.ylabel('Partial Autocorrelation')
plt.show()
```

### **3.7. –≠—Ç–∞–ø 7. –î–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞**

**–ó–∞–¥–∞—á–∞:** –†–∞–∑–¥–µ–ª–∏—Ç—å —Ä—è–¥ –Ω–∞ —Å–æ—Å—Ç–∞–≤–ª—è—é—â–∏–µ ‚Äî —á—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å, —á—Ç–æ ‚Äú–¥–≤–∏–∂–µ—Ç‚Äù –¥–∞–Ω–Ω—ã–º–∏.

#### üß© –ß—Ç–æ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å:
- –ü—Ä–∏–º–µ–Ω–∏—Ç—å –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—é:
  - –ê–¥–¥–∏—Ç–∏–≤–Ω—É—é: `value = trend + seasonality + residual`
  - –ú—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω—É—é: `value = trend √ó seasonality √ó residual` ‚Äî –µ—Å–ª–∏ –∞–º–ø–ª–∏—Ç—É–¥–∞ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ —Ä–∞—Å—Ç—ë—Ç —Å–æ –≤—Ä–µ–º–µ–Ω–µ–º
- –í–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∫–∞–∂–¥—É—é –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—É –æ—Ç–¥–µ–ª—å–Ω–æ
- –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å:
  - –§–æ—Ä–º—É –∏ —Å–∏–ª—É —Ç—Ä–µ–Ω–¥–∞ (–ª–∏–Ω–µ–π–Ω—ã–π? —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–π?)
  - –ü–µ—Ä–∏–æ–¥–∏—á–Ω–æ—Å—Ç—å –∏ –∞–º–ø–ª–∏—Ç—É–¥—É —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ (–Ω–µ–¥–µ–ª—å–Ω–∞—è? –º–µ—Å—è—á–Ω–∞—è? –≥–æ–¥–æ–≤–∞—è?)
  - –ü–æ–≤–µ–¥–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤ ‚Äî —Å–ª—É—á–∞–π–Ω—ã –ª–∏ –æ–Ω–∏? –ï—Å—Ç—å –ª–∏ –≤ –Ω–∏—Ö —Å—Ç—Ä—É–∫—Ç—É—Ä–∞? (–µ—Å–ª–∏ –µ—Å—Ç—å ‚Äî –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è –Ω–µ–ø–æ–ª–Ω–∞—è)

> üõ†Ô∏è **–ò–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç—ã:** `statsmodels.tsa.seasonal.seasonal_decompose`

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞—è, —á—Ç–æ '–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª' —è–≤–ª—è–µ—Ç—Å—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π
target_col = '–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –Ω–∞—Å–µ–ª–µ–Ω–∏—è, —á–µ–ª'

# –í—ã–ø–æ–ª–Ω—è–µ–º –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ
decomposition_additive = seasonal_decompose(df_weather[target_col], model='additive')

# –ì—Ä–∞—Ñ–∏–∫ additive decomposition (–∞–¥–¥–∏—Ç–∏–≤–Ω–æ–≥–æ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏—è)
fig_additive = decomposition_additive.plot()
fig_additive.set_size_inches(10, 8)
fig_additive.suptitle('Additive Decomposition', y=1.02)
plt.show()

# –í—ã–ø–æ–ª–Ω—è–µ–º –º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω—É—é –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—é
# –ü—Ä–∏–º–µ—á–∞–Ω–∏–µ: –¥–ª—è –º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–æ–π –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ —Ç—Ä–µ–±—É—é—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.
# –ü–µ—Ä–µ–¥ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º –º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–æ–π –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ –ø—Ä–æ–≤–µ—Ä—è–µ–º, –≤—Å–µ –ª–∏ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Ü–µ–ª–µ–≤–æ–º —Å—Ç–æ–ª–±—Ü–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ.
if (df_weather[target_col] > 0).all():
    decomposition_multiplicative = seasonal_decompose(df_weather[target_col], model='multiplicative')

    # –°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–æ–π –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏–∏
    fig_multiplicative = decomposition_multiplicative.plot()
    fig_multiplicative.set_size_inches(10, 8)
    fig_multiplicative.suptitle('Multiplicative Decomposition', y=1.02)
    plt.show()
else:
    print(f"Multiplicative decomposition –ø—Ä–æ–ø—É—â–µ–Ω–æ –∫–∞–∫ '{target_col}' —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–µ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.")
```

### ‚úÖ **3.8. –≠—Ç–∞–ø 8. –†–∞–∑—Ä–∞–±–æ—Ç–∫–∞ –≤–µ–±-–∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å–∞ –¥–ª—è –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞**

**–ó–∞–¥–∞—á–∞:** –°–æ–∑–¥–∞—Ç—å –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω–æ–µ –≤–µ–±-–ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ, –ø–æ–∑–≤–æ–ª—è—é—â–µ–µ –∑–∞–≥—Ä—É–∂–∞—Ç—å –≤—Ä–µ–º–µ–Ω–Ω–æ–π —Ä—è–¥, –≤—ã–±–∏—Ä–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∞–Ω–∞–ª–∏–∑–∞ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∫–ª—é—á–µ–≤—ã–µ –¥–∏–∞–≥–Ω–æ—Å—Ç–∏–∫–∏ –≤ —Ä–µ–∞–ª—å–Ω–æ–º –≤—Ä–µ–º–µ–Ω–∏.

#### üéõÔ∏è –§—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω—ã–µ —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è:
- **–ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞** ‚Äî —á–µ—Ä–µ–∑ —Ñ–∞–π–ª (CSV/Parquet) –∏–ª–∏ –≤—ã–±–æ—Ä –∏–∑ –ø—Ä–µ–¥–∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö –ø—Ä–∏–º–µ—Ä–æ–≤
- **–í—ã–±–æ—Ä —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤** ‚Äî –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –∏ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
- **–ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è:**
  - –í—ã–±–æ—Ä –ø–µ—Ä–∏–æ–¥–∞ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ –¥–ª—è –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ (7, 30, 365 –∏ –¥—Ä.)
  - –í—ã–±–æ—Ä –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ª–∞–≥–æ–≤ –¥–ª—è ACF/PACF
  - –í—ã–±–æ—Ä –æ–∫–Ω–∞ –¥–ª—è —Å–∫–æ–ª—å–∑—è—â–µ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ
  - –ü–µ—Ä–µ–∫–ª—é—á–µ–Ω–∏–µ –º–µ–∂–¥—É –∞–¥–¥–∏—Ç–∏–≤–Ω–æ–π/–º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–æ–π –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏–µ–π
- **–ì–µ–Ω–µ—Ä–∞—Ü–∏—è –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–≥–æ HTML-–æ—Ç—á—ë—Ç–∞ —Å:**
  - –ì—Ä–∞—Ñ–∏–∫–æ–º –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞ —Å —Ç—Ä–µ–Ω–¥–æ–º –∏ —Å–∫–æ–ª—å–∑—è—â–∏–º —Å—Ä–µ–¥–Ω–∏–º
  - Heatmap –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
  - –ì—Ä–∞—Ñ–∏–∫–∞–º–∏ ACF –∏ PACF (—Å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å—é –∏–∑–º–µ–Ω–µ–Ω–∏—è max_lag)
  - –ì—Ä–∞—Ñ–∏–∫–∞–º–∏ –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏–∏ (—Ç—Ä–µ–Ω–¥, —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç—å, –æ—Å—Ç–∞—Ç–∫–∏)
  - –†–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ —Ç–µ—Å—Ç–æ–≤ –Ω–∞ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ—Å—Ç—å (ADF, KPSS)
- **–≠–∫—Å–ø–æ—Ä—Ç –æ—Ç—á—ë—Ç–∞** –≤ HTML –∏–ª–∏ PDF

#### üß∞ –¢–µ—Ö–Ω–æ–ª–æ–≥–∏—á–µ—Å–∫–∏–π —Å—Ç–µ–∫ (–Ω–∞ –≤—ã–±–æ—Ä):

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller, kpss
import matplotlib.pyplot as plt
import io

st.set_page_config(page_title="–ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–æ–≤", layout="wide")
st.title("–ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–æ–≤")

uploaded_file = st.file_uploader("–ó–∞–≥—Ä—É–∑–∏—Ç–µ –≤–∞—à —Ñ–∞–π–ª —Å –≤—Ä–µ–º–µ–Ω–Ω—ã–º–∏ —Ä—è–¥–∞–º–∏ (CSV)", type="csv")

if uploaded_file is not None:
    try:
        df = pd.read_csv(uploaded_file)
    except Exception as e:
        st.error(f"–ù–µ —É–¥–∞–ª–æ—Å—å –ø—Ä–æ—á–∏—Ç–∞—Ç—å CSV: {e}")
        st.stop()

    st.write("–î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã (–ø–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏):")
    st.dataframe(df.head())

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –≤—Ä–µ–º–µ–Ω–Ω–æ–π –º–µ—Ç–∫–æ–π –≤ datetime –∏ —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏–Ω–¥–µ–∫—Å–∞
    try:
        df['timestamp'] = pd.to_datetime(df.iloc[:, 0])
        df = df.set_index('timestamp')
        df = df.sort_index()
    except Exception as e:
        st.error(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏ –≤—Ä–µ–º–µ–Ω–Ω–æ–π –º–µ—Ç–∫–∏: {e}")
        st.stop()

    st.write("–î–∞–Ω–Ω—ã–µ —Å –∏–Ω–¥–µ–∫—Å–æ–º –≤—Ä–µ–º–µ–Ω–∏:")
    st.dataframe(df.head())

    # –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ —á–∏—Å–ª–æ–≤—ã—Ö –∫–æ–ª–æ–Ω–æ–∫: –ø–æ–ø—ã—Ç–∫–∞ –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å
    for col in df.columns:
        # –µ—Å–ª–∏ –∫–æ–ª–æ–Ω–∫–∞ –≤—ã–≥–ª—è–¥–∏—Ç —á–∏—Å–ª–æ–≤–æ–π, –ø–æ–ø—Ä–æ–±—É–µ–º –ø—Ä–∏–≤–µ—Å—Ç–∏
        if df[col].dtype == object:
            try:
                df[col] = pd.to_numeric(df[col].str.replace(',', '.').str.replace('¬∞','').str.replace('+',''), errors='ignore')
            except Exception:
                pass

    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
    if len(numerical_cols) == 0:
        st.error("–í –¥–∞—Ç–∞—Å–µ—Ç–µ –Ω–µ—Ç —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞.")
        st.stop()

    target_col = st.selectbox("–í—ã–±–µ—Ä–∏—Ç–µ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é:", numerical_cols)

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞
    st.header("–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞")
    fig_ts = px.line(df, y=target_col)
    st.plotly_chart(fig_ts, use_container_width=True)

    # –î–µ—Å–∫—Ä–∏–ø—Ç–∏–≤–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
    with st.expander("–î–µ—Å–∫—Ä–∏–ø—Ç–∏–≤–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞"):
        st.write(df[numerical_cols].describe())



    import plotly.express as px

    st.header("–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π")
    corr_method = st.selectbox("–ú–µ—Ç–æ–¥ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:", ["pearson", "spearman"])

# –ü—Ä–∞–≤–∏–ª—å–Ω–∞—è —Ü–≤–µ—Ç–æ–≤–∞—è —Å—Ö–µ–º–∞ –¥–ª—è Plotly

 fig_corr = px.imshow(
    df[numerical_cols].corr(),
    text_auto=True,
    aspect="auto",
    color_continuous_scale=px.colors.diverging.RdBu  # –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–æ
)



    # ACF –∏ PACF –≥—Ä–∞—Ñ–∏–∫–∏
    st.header("ACF –∏ PACF –≥—Ä–∞—Ñ–∏–∫–∏")
    lags = st.slider("–í—ã–±–µ—Ä–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–≥–æ–≤ –¥–ª—è ACF/PACF:", 1, 200, 40)

    fig_acf, ax_acf = plt.subplots(figsize=(10, 4))
    try:
        plot_acf(df[target_col].dropna(), lags=lags, ax=ax_acf)
        st.pyplot(fig_acf)
    except Exception as e:
        st.warning(f"–ù–µ —É–¥–∞–ª–æ—Å—å –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å ACF: {e}")

    fig_pacf, ax_pacf = plt.subplots(figsize=(10, 4))
    try:
        plot_pacf(df[target_col].dropna(), lags=lags, ax=ax_pacf, method='ywm')
        st.pyplot(fig_pacf)
    except Exception as e:
        st.warning(f"–ù–µ —É–¥–∞–ª–æ—Å—å –Ω–∞—Ä–∏—Å–æ–≤–∞—Ç—å PACF: {e}")

    # –î–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞
    st.header("–î–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞")
    period = st.number_input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–∏–æ–¥ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 7 –¥–ª—è –Ω–µ–¥–µ–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö):", min_value=1, value=7)

    try:
        decomposition_additive = seasonal_decompose(df[target_col].dropna(), model='additive', period=period)
        fig_decomp_add = decomposition_additive.plot()
        fig_decomp_add.suptitle('–ê–¥–¥–∏—Ç–∏–≤–Ω–∞—è –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è', y=1.02)
        st.pyplot(fig_decomp_add)
    except Exception as e:
        st.warning(f"–ù–µ —É–¥–∞–ª–æ—Å—å –≤—ã–ø–æ–ª–Ω–∏—Ç—å –∞–¥–¥–∏—Ç–∏–≤–Ω—É—é –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—é: {e}")

    try:
        if (df[target_col].dropna() > 0).all():
             decomposition_multiplicative = seasonal_decompose(df[target_col].dropna(), model='multiplicative', period=period)
             fig_decomp_mul = decomposition_multiplicative.plot()
             fig_decomp_mul.suptitle('–ú—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–∞—è –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è', y=1.02)
             st.pyplot(fig_decomp_mul)
        else:
             st.info("–ú—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–∞—è –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è –ø—Ä–æ–ø—É—â–µ–Ω–∞ (–µ—Å—Ç—å –Ω–µ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è).")
    except Exception as e:
        st.warning(f"–ù–µ —É–¥–∞–ª–æ—Å—å –≤—ã–ø–æ–ª–Ω–∏—Ç—å –º—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω—É—é –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—é: {e}")

    # –¢–µ—Å—Ç—ã –Ω–∞ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ—Å—Ç—å (ADF –∏ KPSS)
    st.header("–¢–µ—Å—Ç—ã –Ω–∞ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ—Å—Ç—å")

    st.subheader("–¢–µ—Å—Ç –î–∏–∫–∏-–§—É–ª–ª–µ—Ä–∞ (ADF)")
    try:
        adf_result = adfuller(df[target_col].dropna())
        st.write(f"ADF Statistic: {adf_result[0]}")
        st.write(f"p-value: {adf_result[1]}")
        st.write("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
        for key, value in adf_result[4].items():
            st.write(f"  {key}: {value}")
        if adf_result[1] <= 0.05:
            st.success("–í—ã–≤–æ–¥: –°–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —Ä—è–¥ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")
        else:
            st.info("–í—ã–≤–æ–¥: C–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —Ä—è–¥ –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")
    except Exception as e:
        st.warning(f"ADF —Ç–µ—Å—Ç –Ω–µ –≤—ã–ø–æ–ª–Ω–∏–ª—Å—è: {e}")

    st.subheader("–¢–µ—Å—Ç –ö–≤—è—Ç–∫–æ–≤—Å–∫–æ–≥–æ-–§–∏–ª–ª–∏–ø—Å–∞-–®–º–∏–¥—Ç–∞-–®–∏–Ω–∞ (KPSS)")
    try:
        kpss_result = kpss(df[target_col].dropna(), regression='c')
        st.write(f"KPSS Statistic: {kpss_result[0]}")
        st.write(f"p-value: {kpss_result[1]}")
        st.write("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
        for key, value in kpss_result[3].items():
            st.write(f"  {key}: {value}")
        if kpss_result[1] >= 0.05:
            st.success("–í—ã–≤–æ–¥: –†—è–¥, —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω—ã–º.")
        else:
            st.info("–í—ã–≤–æ–¥: –†—è–¥, —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")
    except Exception as e:
        st.warning(f"KPSS —Ç–µ—Å—Ç –Ω–µ –≤—ã–ø–æ–ª–Ω–∏–ª—Å—è: {e}")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller, kpss
import matplotlib.pyplot as plt
import io

st.title("–ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–æ–≤")

uploaded_file = st.file_uploader("–ó–∞–≥—Ä—É–∑–∏—Ç–µ –≤–∞—à —Ñ–∞–π–ª —Å –≤—Ä–µ–º–µ–Ω–Ω—ã–º–∏ —Ä—è–¥–∞–º–∏ (CSV)", type="csv")

if uploaded_file is not None:
    df = pd.read_csv(uploaded_file)
    st.write("–î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã:")
    st.write(df.head())

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ —Å –≤—Ä–µ–º–µ–Ω–Ω–æ–π –º–µ—Ç–∫–æ–π –≤ datetime –∏ —É—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∏–Ω–¥–µ–∫—Å–∞
    # –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç—Å—è, —á—Ç–æ –ø–µ—Ä–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —è–≤–ª—è–µ—Ç—Å—è –≤—Ä–µ–º–µ–Ω–Ω–æ–π –º–µ—Ç–∫–æ–π
    try:
        df['timestamp'] = pd.to_datetime(df.iloc[:, 0])
        df = df.set_index('timestamp')
        df = df.sort_index()
    except Exception as e:
        st.error(f"–û—à–∏–±–∫–∞ –ø—Ä–∏ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–∏ –≤—Ä–µ–º–µ–Ω–Ω–æ–π –º–µ—Ç–∫–∏: {e}")
        st.stop()

    st.write("–î–∞–Ω–Ω—ã–µ —Å –∏–Ω–¥–µ–∫—Å–æ–º –≤—Ä–µ–º–µ–Ω–∏:")
    st.write(df.head())

    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
    if len(numerical_cols) == 0:
        st.error("–í –¥–∞—Ç–∞—Å–µ—Ç–µ –Ω–µ—Ç —á–∏—Å–ª–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞.")
        st.stop()

    target_col = st.selectbox("–í—ã–±–µ—Ä–∏—Ç–µ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é:", numerical_cols)

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞
    st.header("–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞")
    fig_ts = px.line(df, y=target_col)
    st.plotly_chart(fig_ts)

    # –î–µ—Å–∫—Ä–∏–ø—Ç–∏–≤–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
    st.header("–î–µ—Å–∫—Ä–∏–ø—Ç–∏–≤–Ω–∞—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞")
    st.write(df[numerical_cols].describe())

    # –ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π
st.header("–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π")
fig_corr = px.imshow(
    df[numerical_cols].corr(),
    text_auto=True,
    aspect="auto",
    color_continuous_scale=px.colors.diverging.RdBu  # <- –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–æ)
st.plotly_chart(fig_corr, use_container_width=True))

    # ACF –∏ PACF –≥—Ä–∞—Ñ–∏–∫–∏
    st.header("ACF –∏ PACF –≥—Ä–∞—Ñ–∏–∫–∏")
    lags = st.slider("–í—ã–±–µ—Ä–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∞–≥–æ–≤ –¥–ª—è ACF/PACF:", 1, 100, 40)

    fig_acf, ax_acf = plt.subplots(figsize=(12, 6))
    plot_acf(df[target_col].dropna(), lags=lags, ax=ax_acf)
    st.pyplot(fig_acf)

    fig_pacf, ax_pacf = plt.subplots(figsize=(12, 6))
    plot_pacf(df[target_col].dropna(), lags=lags, ax=ax_pacf)
    st.pyplot(fig_pacf)


    # –î–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞
    st.header("–î–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞")
    period = st.number_input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–∏–æ–¥ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, 7 –¥–ª—è –Ω–µ–¥–µ–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö):", min_value=1, value=7)

    try:
        decomposition_additive = seasonal_decompose(df[target_col].dropna(), model='additive', period=period)
        fig_decomp_add = decomposition_additive.plot()
        fig_decomp_add.suptitle('–ê–¥–¥–∏—Ç–∏–≤–Ω–∞—è –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è', y=1.02)
        st.pyplot(fig_decomp_add)

        if (df[target_col].dropna() > 0).all():
             decomposition_multiplicative = seasonal_decompose(df[target_col].dropna(), model='multiplicative', period=period)
             fig_decomp_mul = decomposition_multiplicative.plot()
             fig_decomp_mul.suptitle('–ú—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–∞—è –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è', y=1.02)
             st.pyplot(fig_decomp_mul)
        else:
             st.write("–ú—É–ª—å—Ç–∏–ø–ª–∏–∫–∞—Ç–∏–≤–Ω–∞—è –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—è –ø—Ä–æ–ø—É—â–µ–Ω–∞, —Ç–∞–∫ –∫–∞–∫ —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–µ–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.")

    except Exception as e:
        st.warning(f"–ù–µ —É–¥–∞–ª–æ—Å—å –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Å–µ–∑–æ–Ω–Ω—É—é –¥–µ–∫–æ–º–ø–æ–∑–∏—Ü–∏—é —Å –≤—ã–±—Ä–∞–Ω–Ω—ã–º –ø–µ—Ä–∏–æ–¥–æ–º: {e}")


    # –¢–µ—Å—Ç—ã –Ω–∞ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ—Å—Ç—å (ADF –∏ KPSS)
    st.header("–¢–µ—Å—Ç—ã –Ω–∞ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω–æ—Å—Ç—å")

    st.subheader("–¢–µ—Å—Ç –î–∏–∫–∏-–§—É–ª–ª–µ—Ä–∞ (ADF)")
    adf_result = adfuller(df[target_col].dropna())
    st.write(f"ADF Statistic: {adf_result[0]}")
    st.write(f"p-value: {adf_result[1]}")
    st.write("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
    for key, value in adf_result[4].items():
        st.write(f"  {key}: {value}")
    if adf_result[1] <= 0.05:
        st.write("–í—ã–≤–æ–¥: –°–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —Ä—è–¥ —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")
    else:
        st.write("–í—ã–≤–æ–¥: C–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —Ä—è–¥ –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")

    st.subheader("–¢–µ—Å—Ç –ö–≤—è—Ç–∫–æ–≤—Å–∫–æ–≥–æ-–§–∏–ª–ª–∏–ø—Å–∞-–®–º–∏–¥—Ç–∞-–®–∏–Ω–∞ (KPSS)")
    try:
        kpss_result = kpss(df[target_col].dropna(), regression='c')
        st.write(f"KPSS Statistic: {kpss_result[0]}")
        st.write(f"p-value: {kpss_result[1]}")
        st.write("–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
        for key, value in kpss_result[3].items():
            st.write(f"  {key}: {value}")
        if kpss_result[1] >= 0.05:
            st.write("–í—ã–≤–æ–¥: –†—è–¥, —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–Ω—ã–º.")
        else:
            st.write("–í—ã–≤–æ–¥: –†—è–¥, —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ, –Ω–µ—Å—Ç–∞—Ü–∏–æ–Ω–∞—Ä–µ–Ω.")
    except Exception as e:
        st.warning(f"–ù–µ —É–¥–∞–ª–æ—Å—å –≤—ã–ø–æ–ª–Ω–∏—Ç—å —Ç–µ—Å—Ç KPSS: {e}")
```
